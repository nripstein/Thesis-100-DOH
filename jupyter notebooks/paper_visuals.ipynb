```python
from bounding_box import bounding_box as bb
import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import ast
%load_ext autoreload
%autoreload 2
```

```python
def draw_bbox(img, coords, label=None, color=None) -> np.ndarray:
    bb.add(img, coords[0], coords[1], coords[2], coords[3], label=label, color=color)
    return img
```

```python
nr_dir = '/home/nripstein/Documents/thesis data/thesis labels/nr labels/all_nr/'
sr_dir = '/home/nripstein/Documents/thesis data/thesis labels/sr labels/'
sv_dir = '/home/nripstein/Documents/thesis data/thesis labels/sv labels/'
```

```python
bboxes = [[206, 150, 354, 308], # r
          [109, 158, 215, 302], # l
          [192, 3, 315, 105], # exper
          [208, 70, 275, 146]] # obj

img = cv2.imread(nr_dir + "nr_shrunk/1591_nr.jpg")

img = draw_bbox(img, bboxes[3], "Object", "aqua")
img = draw_bbox(img, bboxes[0], "Right", "black")
img = draw_bbox(img, bboxes[1], "Left", "black")
img = draw_bbox(img, bboxes[2], label="Experimenter", color="silver")


plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.savefig("1591_nr_det.png", dpi=300)
plt.show()
```
